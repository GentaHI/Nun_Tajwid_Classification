```python
def get_audio_durations(data_path):
    # Get the list of all folders in your data directory
    labels = os.listdir(data_path)

    durations = []

    # Loop over each label (folder)
    for label in labels:
        print(label)
        # Get the list of all audio files in this folder
        files = glob.glob(os.path.join(data_path, label, "*.wav"))

        # For each file, load with librosa and get duration
        for file in files:
            y, sr = librosa.load(file)
            duration = librosa.get_duration(y=y, sr=sr)
            durations.append(duration)

    return durations
```

#Simpan MFCC

```python
data_path = "data"
json_path = "data_mfcc.json"


def save_mfcc(dataset_path, json_path, num_mfcc=13, n_fft=2048, hop_length=256):
  #kamus untuk maping sama label
  data = {
        "mapping": [],
        "labels": [],
        "mfcc": []
        }

  for i, (dirpath, dirnames, filenames) in enumerate(os.walk(dataset_path)):

        # Memastikan sedang memroses sub-folder “data” yaitu sedang memroses folder ikhfa, idzhar, iqlab, dan idghom
        if dirpath is not dataset_path:
          semantic_label = dirpath.split("/")[-1]
          data["mapping"].append(semantic_label)
          print("\nProcessing: {}".format(semantic_label))

          for f in filenames:
            file_path = os.path.join(dirpath, f)
            signal, sample_rate = librosa.load(file_path)

            # perlu zero padding menjadi 3 detik
            padded_audio = librosa.util.fix_length(signal, size=3*sample_rate)

            mfcc = librosa.feature.mfcc(y=padded_audio, sr=sample_rate, n_mfcc=num_mfcc, n_fft=n_fft, hop_length=hop_length)
            mfcc = mfcc.T

            data["mfcc"].append(mfcc.tolist())
            data["labels"].append(i-1)
            print("{}, label ke-:{}".format(file_path, i))

  with open(json_path, "w") as fp:
        json.dump(data, fp, indent=4)
```
